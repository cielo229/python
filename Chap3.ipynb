```python
score=[90,100,88,60,79]
print(score[0])
print(score[1])
for i in range(len(score)):  #len(scare): scare的長度
  print(score[i])
```

```python
score=[90,100,88,60,79]
print(score[-1])
print(score[-2])
print(score[0:3]) #取出score[0]-score[2],不包含score[3]
print(score[:]) #取出所有值
print(score[:4]) #取出score[0]-score[3],不包含score[4]
print(score[2:]) #取出score[2]後面所有的值
print(score[0:5:2]) #間隔2
print(score[::2]) #間隔2
print(score[::-2]) #間隔-2,代表從列表最後往前取值,間隔2
print(score[4:0:-2])
```

```python
my_list0=[]
print(my_list0)
my_list1=list()
my_list2=['python','javascript','C#']
print(my_list2)
print(len(my_list2)) #len:計算list的長度
print(my_list2[2])
my_list2.append('java') #append:附加在最後面
print(my_list2)
my_list2.remove('java') #remove:移除
print(my_list2)
my_list2.insert(0,'java') #insert:插入物件,需要指定索引位置
print(my_list2)
my_list2.clear() #clear:清除所有物件
print(my_list2)
```

```python
a=[1,2,3,4,5,6,7,8,9]
del a[7:9] #刪除某範圍值, 7, 8
print(a)
print(min(a)) #最小值
print(max(a)) #最大值
print(a.index(1)) #找出1所在的第一個索引值
b=[1,0,0,1,1,1]
print(b.count(1)) #1出現的次數
a.reverse() #串列反轉
print(a)
a.sort() #串列由小到大排序,a的值會被改變
print(a)
```
