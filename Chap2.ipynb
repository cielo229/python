#Chap2 控制流程

```python
a=3
b=4
if a>b :
  print("a>b")
elif a==b:
  print("a=b")
else:
  print("b>a")
```

```python
a=int(input("請輸入數字a:"))  #float小數,eval計算
b=int(input("請輸入數字b:"))
if a>b :
  print("a>b")
elif a==b:
  print("a=b")
else:
  print("b>a")
```

```python
print("成績評鑑:")
grade=int(input("請輸入成績:"))
if grade>=90:
  print("優等")
elif grade>=80:
  print("甲等")
elif grade>=70:
  print("乙等")
elif grade>=60:
  print("丙等")
else:
  print("不及格")
```

##for迴圈

```python
for i in range(10): #range(10):0,1,2,3,.....9
  print(i)
```

```python
for i in range(0,10,1): #range(0開始,9結束(不包含10)每次進1步)
  print(i)
```

```python
sum=0
for count in range(1,11,1):
  sum=sum+count
print("1+2+3+...+10=",sum)
```

```python
for a in "I am a cat":
  print(a)
```

```python
for score in [90,100,88,77]:
  print(score)
```
