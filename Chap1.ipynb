### 運算式

```python
a=5
b=3
#透過第三變數c,交換a,b的值
c=a
a=b
b=c
print(a)
print(b)
```

```python
#不透過第三變數,交換a,b的值
a=5
b=3
a=a+b
b=a-b
a=a-b
print(a)
print(b)
```

```python
a=5
b=3
a,b=b,a
print(a)
print(b)
```

## print的三種方式
1.透過,區隔要列印的變數 ex:print("This","is","a","book")
2.利用字串的str.format()(新式)
3.%格式化輸出字串(舊式)

```python
print("This"+"is"+"a"+"book")
print("This","is","a","book")
```

*  利用字串str.format(),在{}中替代變數的位置
*  ex:print("This {} a book".format("is"))
*  {}中可以設定格式
*  例如:{0:10}(10個空格,靠左對齊)
*  例如:{:f}(三個數字用,分隔,小數點後2位數)

```python
print("This is a book".format("is"))
x=123456.7819
print("{:.2f}".format(x))  #小數點.後兩位
print("{:,.2f}".format(x))  #,小數點前用,分開
```

*   C語言類似的%格式
*   %s(字串)%b(整數)%f(浮點數)
*   %s(字串):%10(10格靠右對齊,不足的左方補空白),%-10s
*   %f(浮點數):%7.3f(總共7個浮點數位數,含一個小數點,以及三個小數點後數字)

```python
print("This %10s a book%-10s"%("is","!"))
print("%7.3f"%(687.4333))
```
